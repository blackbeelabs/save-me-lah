```python
from functools import partial
import json
from os import path
import torch

from torch.utils.data import DataLoader
from ruamel.yaml import YAML
import pendulum

import pandas as pd
```

```python
def _get_project_dir_folder():
        return (path.dirname(path.dirname(path.abspath("__FILE__"))))

def get_device():
    if torch.backends.mps.is_available():
        return torch.device("mps")
    elif torch.cuda.is_available():
        return torch.device("cuda")
    else:
        return torch.device("cpu")
```

```python
def main():

    now = pendulum.now()
    model_timestamp = now.format("YYYYMMDDTHHmmss")
    torch.manual_seed(42)
    device = get_device()

    experiment = "baseline"
    ASSETS_FP = path.join(_get_project_dir_folder(), "assets")

    train_folderpath = path.join(
        ASSETS_FP,
        "datasets",
        "CMAPSSData"
    )
    file1_filepath = path.join(train_folderpath, "test_FD001.txt")

    model_fp = path.join(
        ASSETS_FP,
        "models",
        f"model-{experiment}-{model_timestamp}.pkl",
    )
    model_json_fp = path.join(
        ASSETS_FP,
        "models",
        f"modelprofile-{experiment}-{model_timestamp}.json",
    )

    # Signal
    df = pd.read_csv(file1_filepath, sep=" ")
    display(df)
```

```python
main()
```
